```python
import random as ran
```

```python
# this function only use ASCII codes 

def incode(t):
  list = []
  for char in t:
    list.append(ord(char) + 10)
  return list
```

```python
def decode(list):
  str = ""
  for item in list:
    str += chr(item - 10)
  return str
```
